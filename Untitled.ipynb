```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("logs.csv", parse_dates=["Fecha"])
df.info()
```

```python
# Formato: mysql+mysqlconnector://usuario:contraseña@host:puerto/nombre_bd
engine = create_engine("mysql+mysqlconnector://root:root@localhost:3306/logs")
```

```python
df.to_sql(name="logs", con=engine, if_exists="replace", index=False)
```

```python
pd.read_sql("SELECT * FROM logs LIMIT 5", con=engine)
```
